```python
import pandas as pd
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('data/indian-liver-patient.csv')
df
```

```python
age = (df.iloc[:,0])
gender = (df.iloc[:,1])
diagnose= (df.iloc[:,10])
```

```python
sns.distplot(age, 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4});
```

```python
women = df.loc[(df['gender'] == 'Female')]
```

```python
men = df.loc[(df['gender'] == 'Male')]
```

```python
df.plot.bar(x = 'Gender', y = 'Freq', 
                 legend = None, 
                 edgecolor = 'k',
                 rot = 0);
```

```python
sns.distplot(diagnose,         
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4});         
plt.xlabel('Diagnosis')
plt.xticks([1, 2])
plt.show()
```

```python
diagnosis_men =  men.iloc[:,10]
sns.distplot(diagnosis_men, 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4});                           
plt.xlabel('Diagnosis in Men')
plt.xticks([1, 2])
plt.ylim(0, 3.5)
plt.show()
Counter(diagnosis_men)
```

```python
diagnosis_women =  women.iloc[:,10]
sns.distplot(diagnosis_women, 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4});  
plt.xlabel('Diagnosis in Women')
plt.ylim(0, 3.5)
plt.xticks([1, 2])
plt.show()
Counter(diagnosis_women)
```
